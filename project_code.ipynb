```python
# Importing libraries and reading the data from drive
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/df_overall.csv")
```

```python
# Calculating a new variable 'services per 1k'
df['services per 1k'] = (df['TELEHEALTH SERVICE COUNT'] / df['ACS_TOT_POP_WT'] ) * 1000
```

```python
# Dropping the uncessary columns
df = df.drop(['TELEHEALTH SPENDING', 'SERVICE MONTH','SERVICE YEAR', 'PAYER TYPE','TELEHEALTH SERVICE COUNT', 'COUNTY', 'MEMBER POPULATION', 'PROVIDER TYPE', 'DIAGNOSIS TYPE', 'TELEHEALTH DENOMINATOR MEMBER MONTHS', 'IN PERSON DENOMINATOR MEMBER MONTHS', 'IN PERSON SERVICE COUNT', 'ACS_TOT_POP_WT', 'IN PERSON SPENDING', 'COUNTYFIPS'], axis=1)
```

```python
df= df.dropna(axis='columns')
```

```python
# Loading PIml library to the code
from piml import Experiment
exp = Experiment(highcode_only=True)
```

# Load and Prepare Data

```python
# Data load
exp.data_loader(data=df)
```

```python
# Prepare dataset by setting target, task type and train-test split
exp.data_prepare(target='services per 1k', task_type='Regression', test_ratio=0.3, random_state=0)
```

# Train intepretable models

```python
## ReLU-DNN: ReLU Deep Neural Networks with L1-Regularization
from piml.models import ReluDNNRegressor

model1 = ReluDNNRegressor(hidden_layer_sizes=(15,15), l1_reg=1e-03,
                        batch_size=500, learning_rate=0.001)
exp.model_train(model=model1, name='ReLU-DNN')
exp.model_diagnose(model="ReLU-DNN", show='accuracy_table')
```

```python
# exp = Experiment()  # Assuming Experiment is a class that needs to be instantiated
model = exp.get_model("ReLU-DNN")
print(model)
```

```python
exp.model_train(model=model, name='ReLU-DNN')
```

```python
test_X, test_y, _ = exp.get_data(test=True)
```

```python
## GAMI-Net: GAM-Neural Networks with Structured Interactions
from piml.models import GAMINetRegressor

model2 = GAMINetRegressor(interact_num=10, loss_threshold=0.01,
                        subnet_size_main_effect=[20],
                        subnet_size_interaction=[20,20])
exp.model_train(model=model2, name='GAMI-Net')
exp.model_diagnose(model="GAMI-Net", show='accuracy_table')
```

```python
## EBM: Explainable Boosting Machine from MS Research
from piml.models import ExplainableBoostingRegressor

model3 = ExplainableBoostingRegressor(interactions=10, random_state=0)
exp.model_train(model=model3, name='EBM')
exp.model_diagnose(model="EBM", show='accuracy_table')
```
